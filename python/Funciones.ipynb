```python
#1
texto = input('Escribe una palabra o una frase').lower()
def conteo(letras):
    contador = 0
    vocals = ['a','e','i','o','u']
    for i in letras:
        if i in vocals:
            contador += 1
    return contador

resultado = conteo(texto)
print('El número de vocales que tiene el texto:', texto,'es', resultado)
```

```python
#3
frase = input('Escribe una palabra o una frase:').lower()

def unicas(x):
    x= x.split()
    uniques = []
    for i in x:
        if i not in uniques:
            uniques.append(i)
    return uniques

resultado = unicas(frase)
print('Las palabras únicas son:', resultado)
```

```python
#4
listnum = [4,12,65,47,88,96,25,33,34,58,12,20,37,48,52,77]

def pares(x):
    parells = []
    for num in x:
        if num % 2 == 0:
            parells.append(num)
    return parells

listnum.sort() #por defecto sale en orden ascendente para hacerlo descente .sort(reverse=True)
print('La lista de número pares es:', pares(listnum))
```

```python
#5

frase = input('Escribe una frase o palabra').lower()

def conteo(x):
    x = x.replace(" ","")
    dictfrase = {}
    for letra in x:
        if letra in dictfrase:
            dictfrase[letra]+=1
        else:
            dictfrase[letra]=1
    return dictfrase

print('El diccionario resultante es:', conteo(frase))
```

Ejercicio 2:
Crea una funció que rebi una llista de paraules com a paràmetre. 
La funció ha de recórrer la llista i retornar una llista amb la paraula més llarga a la primera posició i la paraula més curta a la segona posició. 
Després, mostra per pantalla les dues paraules.

Fes servir aquesta llista per executar la funció:
arbre, codi, helicòpter, opcions, sal, amanida, cotxe, sofà, llapis, pissarra

*Fes l'exercici sense utilitzar funcions de python com max() o min().*

```python
listta = ['arbre', 'codi', 'helicòpter', 'opcions', 'sal', 'amanida', 'cotxe', 'sofà', 'llapis', 'pissarra']

def orden(x):
    mas_larga = x[0]
    mes_peque = x[0]
   
    for i in x[1:]:
        largo = len(i)
        if largo > len(mas_larga):
            mas_larga = i
        elif largo < len(mes_peque):
            mes_peque = i
    return [mas_larga, mes_peque]

resultado = orden(listta)

print(f'La palabra más larga es {resultado[0]}')
print(f'La palabra más larga es {resultado[1]}')
```

Ejercicio 3:

Crea una funció que demani 5 noms d'alumnes i les seves respectives notes de finals de curs.

A mesura que vagis demanant les dades, has d’afegir:
- Al diccionari d’aprovats, les persones que tenen una nota igual o superior a 5.
- Al diccionari de suspesos, la resta d’alumnes.

La funció ha de retornar el diccionari d'aprovats.

Després, mostra per pantalla el diccionari d’aprovats.

Fes servir aquestes dades per executar la funció:
Laura - 8 || Antoni - 10  || Laia - 4  ||  Albert - 2 || Sara - 9

```python
def notas():
    aprobados = {}
    suspensos = {}

    for i in range(5):
        nombre = input('Escribe el nombre del alumno')
        nota = float(input('Escribe la nota final'))
        if nota >= 5:
            aprobados[nombre] = nota
        elif nota < 5:
            suspensos[nombre] = nota
    return aprobados

notas()
```
