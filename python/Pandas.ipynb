En una empresa, és essencial avaluar el rendiment dels venedors per optimitzar les vendes i millorar els resultats. Un indicador clau és comparar les vendes de cada venedor amb la mitjana de l’equip comercial.

Escriu una funció que donat el dataset vendes.csv:

- Agrupi les vendes per venedor i calculi el total de cada un.
- Calculi la mitjana de vendes de tots els venedors.
- Identifiqui els venedors amb un rendiment inferior a la mitjana.
- Retorni una llista amb els noms dels venedors que han tingut vendes inferiors a la mitjana.

Crida la funció amb el dataframe vendes.csv i mostra els venedors que han tingut un rendiment en vendes per sota de la mitjana.


```python
import pandas as pd

df = pd.read_csv(r'c:\Users\herre\OneDrive\Desktop\IT ACADEMY\PYTHON FUNDAMENTOS II\vendes.csv',encoding="latin1",
    sep=";")

df = df.rename(columns={'ï»¿Venedor':'Venedor'})
df.head()
```

```python
def evaluacion(dataframe):
    totales = dataframe.groupby('Venedor')['Import'].sum()
    media = totales.mean().round(2)
    resultado = totales[totales < media]
    return resultado.index.tolist() #devuelve el resultado de la función pero el index, en este caso el nombre de los vendedores
                                    # y lo convierte en lista con tolist()
```

```python
evaluacion(df)
```
